# Exporting WPA Collection Data from Library of Congress JSON API

To pull collection data from the Library of Congress, we used the [Library of Congress JSON API](https://libraryofcongress.github.io/data-exploration/index.html). The API allows for automated extracting and procesing of bulk data relating to Library of Congress collection items. While the API is technically in Beta and is not yet an official version, the API's documentation is very clearly written and allows for the efficient extraction of collection data without needing to write complex queries. Additionally, there are no API keys or authorization needed to use the API; we could use it right away.

In this workflow, we extract several fields of item information from the [Work Projects Administration (WPA) Poster Collection](https://www.loc.gov/collections/works-progress-administration-posters/).

```python
# Import dependencies, including specifications for "PrettyPrinter" to format JSON in user-friendly way.
import json
import pprint
pp = pprint.PrettyPrinter(indent=4)
import requests
import pandas as pd


# We set the HTML request to the URL for the collection appended with "?fo=json," as specified in the
# API documentation. We then print the initial JSON pull of all of the collection item data.
r = requests.get("https://www.loc.gov/collections/works-progress-administration-posters/?fo=json").json()
print(json.dumps(r, indent=2, sort_keys=True))
```

```python
# From the full JSON of the webpage, we can iterate through and print the actual collection items appearing on the
# page by accessing the "results" element of the JSON, as shown below.

for x in r["results"]:
    print(x)
    print("-------------------")
```

Based on how the API is configured, the initial JSON dump only includes items on the first page of results, giving us only 25 results out of more than 900 collection items. We need to iterate through each page of collection items in order to extact information for every item in the collection. The API's documentation helpfully describes how "pagination" (iterating through pages of results) works. We implemented the logic in the below section, while also saving the results into a list ("full_list"), so that we did not need to access the API everytime we wanted this data.

```python
# Set empty list to store the full results.
full_list = []

# As long as there is another page to acess, implement the following:
while True:
    # For every item in the page results, PrettyPrint the item and append the item to the full list.
    for x in r["results"]: 
        pp.pprint(x)
        full_list.append(x)
        print("---------------------------------------------------")
    # As long as there is a next page ("["pagination"]["next"]"), access the next page, get the JSON for the next
    # page, and process each item on that page (printing and appending to full list).
    next_page = r["pagination"]["next"]
    if next_page is not None:
        r = requests.get(next_page).json()
    # Stop the iteration when there are no more pages to iterate through.
    else:
        break
```

We realized that unfortunately, among the "results" pulled from the API, we receive websites and articles on the Library of Congress website that are related to the collection, in addition to the collection items themselves. While this might be helpful if we are researching the collection on the Library of Congress website, for our purposes, we are only concerns with the actual collection items themselves. Looking at the JSON, we see that the "original format" field defines whether the item is a collection item ("photo, print, drawing") or not ("collection," "web page," etc.). Therefore, from the "full_list" of results, we extracted all true collection items and stored them in a "collection_item" list, as shown below:

```python
# Establish two lists to store the results information. While the collection_item list is more important for our
# purposes, we also wanted to see how many non-collection items were extracted, hence the "other_item" list.
collection_item = []
others_item = []

# For each result in the full_list, if the first "original format" element of the result is a "photo, print, drawing"
# item, than append that item to the "collection_item" list, as we know this is an actual collection item. Otherwise,
# add it to the "others_item" list.
for x in full_list:
    orig_format = x["original_format"][0]
    if orig_format == "photo, print, drawing":
        collection_item.append(x)
    else:
        others_item.append(x)

# Print out the lengths of the two lists to show number of collection items and number of non-collection items.     
print("Collection Items: " + str(len(collection_item)))
print("Other Items: " + str(len(others_item)))
```

Now that we have all of the information for all of the collection items stored in one list, we wanted to create sub-lists of specific types of information for each item that would allow us to organize only basic information about each item and incorporate this information into an easy-to-read pandas dataframe as well as a CSV file exported from the dataframe. For each item, we decided to extract title, web URL, LCCN (Library of Congress Control Number, essentially a unique ID for the item), description, and date.

```python
# Establish empty lists for each data attribute we will be collecting.
title_list = []
url_list = []
lccn_list = []
date_list = []
descrip_list = []

# For each item in the "collection_item" list, extract the data attribute based on its element name and append the
# value to its corresponding list.
for x in collection_item: 
    
    # Extracting title
    t = x["title"]
    title_list.append(t)
    
    # Extracting URL
    u = x["url"]
    url_list.append(u)
    
    # Extracting description (needing just the first description, if there are multiple)
    descrip = x["description"][0]
    descrip_list.append(descrip)
    
    # Extracting LCCN. This attribute is part of every collection item URL, so we used string manipulation to isolate
    # out this specific ID.
    lccn = x["url"][25:][:-1]
    lccn_list.append(lccn)
    
    # Extracting date
    d = x["date"]
    date_list.append(d)
```

```python
# Create a pandas dataframe from the created lists of each attribute and display the first five records.

df = pd.DataFrame({"Title": title_list, 
                   "URL": url_list, 
                   "LCCN": lccn_list,
                   "Date": date_list, 
                   "Description": descrip_list
                  })

df.head()
```

```python
# Export a copy of the pandas dataframe as a CSV file.
df.to_csv('LOC_WPA_posters.csv')
```
